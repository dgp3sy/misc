```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

Read in data file containing customer information for market segmentation. The following data contains customerID (primary key), gender, age, annual income in thousands and spending score - which is a mall rating on how friendly the customer is and how likey he or she is to purchase an item at the store (a ranking designated by the retail store)

```python
df = pd.read_csv("Mall_Customers.csv")
print(df.head())
df.insert(5, "IsMale", None) 
```

```python
df.loc[df.Gender == "Male", 'IsMale'] = 1
df.loc[df.Gender == "Female", 'IsMale'] = 0
print(df.head())
```

```python
plt.scatter(x=df["Annual Income (k$)"], y=df["Age"], s=df["Spending Score (1-100)"], alpha=0.5)
plt.ylabel("Age")
plt.xlabel("Annual Income (thousands of dollars)")
plt.title("Market Segmentation Based On Income and Age")
plt.show()
```

```python
plt.scatter(x=df["Age"], y=df["Spending Score (1-100)"], alpha=0.5)
plt.ylabel("Age")
plt.xlabel("Age")
plt.title("Segmentation by Age")
plt.show()
```

```python
plt.scatter(x=df["Annual Income (k$)"], y=df["Spending Score (1-100)"], c=df["IsMale"], alpha=0.5)
plt.ylabel("Age")
plt.xlabel("Annual Income (thousands of dollars)")
plt.title("Segmentation by Income")
plt.show()
```
